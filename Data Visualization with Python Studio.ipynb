```python
# Here is the setup of the dataframe. Feel free to use .info() or .head() to get a better understanding of what is inside the dataframe!
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import style
# setting style for graphs
style.use('ggplot')
plt.rcParams['figure.figsize'] 
df = pd.read_csv("books.csv", error_bad_lines=False)

years = []

for y in df['publication_date']:
    date = y.split("/")
    years.append(date[2])

df["year"] = years
df
```

```python
# Use your Python skills to create a visualization here.
# Number of books published per year.

# df = sb.load_dataset('books.csv') 
count_of_books=df.groupby('year')['bookID'].sum().to_frame().reset_index()
#Creating the column plot 
plt.bar(df['year'],df['bookID'] )
#Adding the aesthetics
ax = plt.gca()
ax.set_xticks(ax.get_xticks()[::10])
plt.title('Number of books published per year')
plt.xlabel('year')
plt.ylabel('count')
plt.show()
```
